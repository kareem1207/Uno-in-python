```python
import import_ipynb
from uno_cards import Cards
from uno_ai import Ai
from random import shuffle,randint
uno_special_cards=['wild card','+4 card','blank card']
uno_special_color_cards=['ðŸš« card','reverse card','+2 card']
uno_cards_colors=["yellow","red","green","blue"]
uno_avoidable_cards=[]
blank_card_tasks=["draw 10+ cards " , "switch cards with other player ","remove all special cards ","remove all numeric cards "]
for i in uno_special_color_cards:
    for j in uno_cards_colors:
        k=j+':'+i
        uno_avoidable_cards.append(k)
uno_color_special_card=uno_avoidable_cards        
uno_avoidable_cards+=uno_special_cards
#takes name
def names(count=2):
    unoPlayerNames=["ai"]
    player_name=input("Enter your name : ")
    player_name=player_name or "player 1"
    unoPlayerNames.append(player_name)
    shuffle(unoPlayerNames)
    return unoPlayerNames,player_name     
#distrubute cards to both ai and player
def cards_distribution(names,count):
    uno_ai_cards=Cards(count)
    uno_player_cards=Cards(count)
    for i in names :
        if i=="ai":
            ai_cards=uno_ai_cards.cards()
            while True:
                if len(ai_cards)==0:
                    ai_cards=uno_ai_cards.cards()
                elif len(ai_cards)==count:
                    break
        elif i!="ai":
            player_cards=uno_player_cards.cards()
            while True:
                if len(player_cards)==0:
                    player_cards=uno_player_cards.cards()
                elif len(player_cards)==count:
                    break
    return ai_cards,player_cards
#Takes first card 
def first_chance(names,ai_cards,player_cards):
    game=True
    name_count=0
    card_on_table=""
    ai=Ai(ai_cards)
    if names[0]=="ai":
        k=ai.ai_game()
        card_on_table=k[1]
    else :
        for i in range(len(player_cards)):
            print(f" key:{i} = assosiated card {player_cards[i]}")
        player_choice=input("Enter your choice :")
        if player_choice.isnumeric():
            card_on_table=player_cards.pop(int(player_choice))
            if card_on_table in uno_avoidable_cards:
                print("Wrong card taking closest numeric card ")
                for i in player_cards:
                    if i not in uno_avoidable_cards:
                        card_on_table=player_cards.pop(player_cards.index(i))
                        break
    return card_on_table,ai_cards,player_cards
#players special cards func decider
def player_special_cards(card_on_table,ai_cards,player_cards):
    if card_on_table in uno_special_cards:
        match(card_on_table):
            case "wild card":
                for i in range(len(uno_cards_colors)):
                    print(f" key:{i} = assosiated card {uno_cards_colors[i]}")
                color_choice=int(input("Enter the color choice according to the key"))
                card_on_table=uno_cards_colors[color_choice]
            case "+4 card":
                for i in range(len(uno_cards_colors)):
                    print(f" key:{i} = assosiated card {uno_cards_colors[i]}")
                color_choice=int(input("Enter the color choice according to the key"))
                card_on_table=uno_cards_colors[color_choice]
                card=Cards(4)
                adding_cards=card.cards()
                ai_cards+=adding_cards
            case "blank card":
                for i in range(len(uno_cards_colors)):
                    print(f" key:{i} = assosiated card {uno_cards_colors[i]}")
                color_choice=int(input("Enter the color choice according to the key(only int)"))
                card_on_table=uno_cards_colors[color_choice]    
                for i in range(len(blank_card_tasks)):
                    print(f" key:{i} = assosiated card {blank_card_tasks[i]}")
                task_choice=int(input("Enter task according to your wish and follow key value (use integers only )"))
                match task_choice :
                    case 0:
                        card=Cards(10)
                        adding_cards=card.cards()
                        ai_cards+=adding_cards
                    case 1 :
                        dummy_card=ai_cards
                        ai_cards=player_cards
                        player_cards=dummy_cards
                    case 2:
                        for i in ai_cards:
                            if i in uno_avoidable_cards:
                                ai_cards.remove(i)
                    case 3 :
                        for i in ai_cards:
                            if i not in uno_avoidable_cards:
                                ai_cards.remove(i)
                    case _ :
                        print("Wrong input so no task is assigned" )
    return card_on_table,ai_cards,player_cards
#ai special cards func decider
def ai_special_cards(card_on_table,ai_cards,player_cards):
    if card_on_table in uno_special_cards:
        match(card_on_table):
            case "wild card":
                color_choice=randint(0,3)
                card_on_table=uno_cards_colors[color_choice]
            case "+4 card":
                color_choice=randint(0,3)
                card_on_table=uno_cards_colors[color_choice]
                card=Cards(4)
                adding_cards=card.cards()
                player_cards+=adding_cards
            case "blank card":
                color_choice=randint(0,3)
                card_on_table=uno_cards_colors[color_choice]    
                task_choice=randint(0,3)
                match task_choice :
                    case 0:
                        card=Cards(10)
                        adding_cards=card.cards()
                        player_cards+=adding_cards
                    case 1 :
                        dummy_card=ai_cards
                        ai_cards=player_cards
                        player_cards=dummy_cards
                    case 2:
                        for i in player_cards:
                            if i in uno_avoidable_cards:
                                player_cards.remove(i)
                    case 3 :
                        for i in player_cards:
                            if i not in uno_avoidable_cards:
                                player_cards.remove(i)
                    case _ :
                        print("Wrong input so no task is assigned" )
    return card_on_table,ai_cards,player_cards
#complete gameplay 
def game_play(name,card_on_table,ai_cards,player_cards):
    game=True
    chance=1
    while game:
        print(f"On table {card_on_table} card is there\n{name[chance]} turn ")
        if name[chance]=="ai":
            ai=Ai(ai_cards,card_on_table)
            ai_choice=ai.ai_game()
            ai_cards,card_on_table=ai_choice[0],ai_choice[2] 
            if card_on_table in uno_avoidable_cards:
                if card_on_table in uno_special_cards:
                    fake=ai_special_cards(card_on_table,ai_cards,player_cards)
                    card_on_table,ai_cards,player_cards=fake[0],fake[1],fake[2]
                elif card_on_table in uno_color_special_card:
                    wrong=card_on_table.split(':')
                    if wrong[1]=='ðŸš« card':
                        chance=0
                    elif wrong[1]=='reverse card':
                        name.reverse()
                    elif wrong[1]=='+2 card':
                        card=Cards(2)
                        adding_card=card.cards()
                        player_cards+=adding_card
            if len(ai_cards)==1:
                print("Uno!")
            elif len(ai_cards)==0:
                print("Ai won the game ")
                game=False
        elif name[chance]!="ai":
            for i in range(len(player_cards)):
                print(f" key:{i} = assosiated card {player_cards[i]}")
            print("press p for picking a card ")
            player_choice=input("Enter your choice :")
            if player_choice.isnumeric():
                card_on_table=player_cards.pop(int(player_choice))
                if card_on_table in uno_avoidable_cards:
                    if card_on_table in uno_special_cards:
                        fake=player_special_cards(card_on_table,ai_cards,player_cards)
                        card_on_table,ai_cards,player_cards=fake[0],fake[1],fake[2]
                    elif card_on_table in uno_color_special_card:
                        wrong=card_on_table.split(':')
                        if wrong[1]=='ðŸš« card':
                            chance=0
                        elif wrong[1]=='reverse card':
                            name.reverse()
                        elif wrong[1]=='+2 card':
                            card=Cards(2)
                            adding_card=card.cards()
                            ai_cards+=adding_card
            elif player_choice=='p':
                card=Cards(1)
                dummy1=card.cards()
                player_cards.append(dummy1[0])
            if len(player_cards)==1:
                print(f"{name[chance]} shouts Uno!")
            elif len(player_cards)==0:
                print(f"{name[chance]} won the game ")
                game=False
        chance+=1
        if chance>=2:
            chance=0
if __name__=="__main__":
    uno_player_names=names()
    print(f"This is the order {uno_player_names[0]}")
    try:
        cards_count=int(input("Enter cards count"))
    except Exception as e :
        cards_count=7
        print(f"Faced {e} error so making {cards_count}'s as default cards count ")
    cards=cards_distribution(uno_player_names[0],cards_count)
    print(f"This are {uno_player_names[1]} cards {cards[1]}")
    dummy=first_chance(uno_player_names[0],cards[0],cards[1])
    game_play(uno_player_names[0],dummy[0],dummy[1],dummy[2])
```
