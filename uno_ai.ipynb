```python
import import_ipynb
from uno_cards import Cards
from random import choice,randint
#converts list to dic 
def list_to_dict(lst):
    result = {}
    special_count = 1 
    colon_keys = set() 
    for item in lst:
        if ':' in item:
            key, value = item.split(':', 1)
            key = key.strip()
            value = value.strip()
            result[key] = value
            colon_keys.add(key)
        else:
            special_key = f"special_card_{special_count}"
            result[special_key] = item.strip()
            special_count += 1
    return result, colon_keys
# reverse the dic 
def reverse_dict_with_condition(d, colon_keys):
    reversed_d = {}
    for key, value in d.items():
        if key in colon_keys:
            reversed_d[value] = key
        else:
            reversed_d[key] = value
    return reversed_d
#ai 
class Ai :
    #selection of cards
    def __init__(self,cards,table_card=""):
        self.ai_cards=cards
        self.card_on_table=table_card
        uno_special_cards=['wild card','+4 card','blank card']
        uno_special_color_cards=['ðŸš« card','reverse card','+2 card']
        uno_cards_colors=["yellow","red","green","blue"]
        uno_avoidable_cards=[]
        for i in uno_special_color_cards:
            for j in uno_cards_colors:
                k=j+':'+i
                uno_avoidable_cards.append(k)
        uno_avoidable_cards+=uno_special_cards
        while True:
            if table_card=="":
                self.card_on_table=choice(self.ai_cards)
                if self.card_on_table not in uno_avoidable_cards :
                    self.ai_cards.pop(self.ai_cards.index(self.card_on_table))
                    break
            else :
                break
        self.card_on_table_dummy=self.card_on_table
    #ai game 
    def ai_game(self):
        uno_special_cards=['wild card','+4 card','blank card']
        c=self.card_on_table.split(':')
        ai_cards_dic=list_to_dict(self.ai_cards)
        ai_cards_copy=ai_cards_dic[0]
        ai_cards_copy2=reverse_dict_with_condition(ai_cards_dic[0],ai_cards_dic[1])
        if c[0] in ai_cards_copy and len(c)==2:
            dummy=c[0]+':'+ai_cards_copy[c[0]]
            self.card_on_table=self.ai_cards.pop(self.ai_cards.index(dummy))
            ai_cards_copy.pop(c[0])
        elif len(c)==2 and c[1] in ai_cards_copy2 :
            dummy=ai_cards_copy2[c[1]]+':'+c[1]
            self.card_on_table=self.ai_cards.pop(self.ai_cards.index(dummy))
            ai_cards_copy2.pop(c[1])
        elif c[0] not in ai_cards_copy:
            for i in uno_special_cards:
                if i in self.ai_cards:
                    self.card_on_table=self.ai_cards.pop(self.ai_cards.index(i))
                    break
        elif c[0] not in self.ai_cards :
            uno_card=Cards(1)
            sub_card=uno_card.cards()
            self.ai_cards.append(sub_card[0])  
        else :
            print("Special card")
            self.card_on_table=self.ai_cards.pop(self.ai_cards.index(c[0]))
        return self.ai_cards,self.card_on_table_dummy,self.card_on_table
if __name__=="__main__":
    uno_card=Cards()
    ai_card=uno_card.cards()
    table_card="red:7"
    ai=Ai(ai_card,table_card)
    for _ in range(0,7):
        j=ai.ai_game()
        print(j[0],j[2])
```
